```python
import numpy as np

A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])

A_inv = np.linalg.inv(A)
result1 = A @ A_inv

print("A * A^-1 =\n", result1)



AB = A @ B
BA = B @ A

print("AB =\n", AB)
print("BA =\n", BA)
print("Are AB and BA equal? →", np.array_equal(AB, BA))



left_side = (AB).T
right_side = B.T @ A.T

print("(AB)^T =\n", left_side)
print("B^T * A^T =\n", right_side)
print("Are they equal? →", np.array_equal(left_side, right_side))

```

```python
import numpy as np

A = np.array([[ 2, -3,  1],
              [ 1, -1,  2],
              [ 3,  1, -1]])

B = np.array([-1, -3, 9])

A_inv = np.linalg.inv(A)
X = A_inv @ B
print("Solution using inverse:", X)

X2 = np.linalg.solve(A, B)
print("Solution using np.linalg.solve:", X2)
```

```python
import time
import numpy as np

# 1. ELEMENT-WISE ADDITION

# Python List addition
list1 = [i for i in range(1_000_000)]
list2 = [i for i in range(1_000_000)]

start = time.time()
list_add = [list1[i] + list2[i] for i in range(1_000_000)]
end = time.time()

print("Python List Addition Time:", end - start, "seconds")

# NumPy addition
arr1 = np.arange(1_000_000)
arr2 = np.arange(1_000_000)

start = time.time()
np_add = arr1 + arr2
end = time.time()

print("NumPy Array Addition Time:", end - start, "seconds")

# 2. ELEMENT-WISE MULTIPLICATION

# Python List multiplication
start = time.time()
list_mul = [list1[i] * list2[i] for i in range(1_000_000)]
end = time.time()

print("\nPython List Multiplication Time:", end - start, "seconds")

# NumPy multiplication
start = time.time()
np_mul = arr1 * arr2
end = time.time()

print("NumPy Array Multiplication Time:", end - start, "seconds")


# 3. DOT PRODUCT

# Python List Dot Product
start = time.time()
dot_list = sum(list1[i] * list2[i] for i in range(1_000_000))
end = time.time()

print("\nPython List Dot Product Time:", end - start, "seconds")

# NumPy Dot Product
start = time.time()
dot_np = np.dot(arr1, arr2)
end = time.time()

print("NumPy Array Dot Product Time:", end - start, "seconds")


# 4. MATRIX MULTIPLICATION (1000 x 1000)

# Python list matrix multiplication
matrix1 = [[1] * 1000 for _ in range(1000)]
matrix2 = [[1] * 1000 for _ in range(1000)]

start = time.time()
result_matrix = [[sum(a * b for a, b in zip(row, col))
                  for col in zip(*matrix2)]
                  for row in matrix1]
end = time.time()

print("\nPython List Matrix Multiplication Time:", end - start, "seconds")


# NumPy matrix multiplication
np_matrix1 = np.ones((1000, 1000))
np_matrix2 = np.ones((1000, 1000))

start = time.time()
np_result = np_matrix1 @ np_matrix2   # or np.dot()
end = time.time()

print("NumPy Array Matrix Multiplication Time:", end - start, "seconds")
```
